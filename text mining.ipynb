```python
import chainer
from chainer import training
from chainer.training import extensions
import chainer.links as L
import chainer.functions as F
from chainer.datasets import tuple_dataset
```

```python
class MLP(chainer.Chain):
    def __init__(self, n_in, n_out):
        super(MLP, self).__init__(
            l1=L.Linear(n_in, 2000),
            l2=L.Linear(2000, 1000),
            l3=L.Linear(1000, 500),
            l4=L.Linear(500, n_out),
        )
    def __call__(self, x):
        h = F.relu(self.l1(x))
        h = F.relu(self.l2(h))
        h = F.relu(self.l3(h))
        h = self.l4(h)
        return h

###### 3層 + BatchNormalization
class MLP_bnorm(chainer.Chain):
    def __init__(self, n_in, n_out):
        super(MLP_bnorm, self).__init__(
            l1=L.Linear(n_in, 1000),
            l2=L.Linear(1000, 1000),
            l3=L.Linear(1000, 500),
            l4=L.Linear(500, n_out),
            bnorm1 = L.BatchNormalization(1000),
            bnorm2 = L.BatchNormalization(1000),
            bnorm3 = L.BatchNormalization(500),
        )
    def __call__(self, x):
        h = F.relu(self.bnorm1(self.l1(x)))
        h = F.relu(self.bnorm2(self.l2(h)))
        h = F.relu(self.bnorm3(self.l3(h)))
        h = self.l4(h)
        return h
        
###### 3層 + BatchNormalization + Dropout
class MLP_bnorm_dropout(chainer.Chain):
    def __init__(self, n_in, n_out):
        super(MLP_bnorm_dropout, self).__init__(
            l1=L.Linear(n_in, 2000),
            l2=L.Linear(2000, 1000),
            l3=L.Linear(1000, 500),
            l4=L.Linear(500, n_out),
            bnorm1 = L.BatchNormalization(2000),
            bnorm2 = L.BatchNormalization(1000),
            bnorm3 = L.BatchNormalization(500),
        )
    def __call__(self, x):
        h = F.dropout(F.relu(self.bnorm1(self.l1(x))), ratio=0.5) #v2からtrain引数は不要なので注意
        h = F.dropout(F.relu(self.bnorm2(self.l2(h))), ratio=0.5)
        h = F.dropout(F.relu(self.bnorm3(self.l3(h))), ratio=0.5)
        h = self.l4(h)
        return h
```

```python
train = tuple_dataset.TupleDataset(x.astype(np.float32), y.astype(np.int32))
test = tuple_dataset.TupleDataset(x_test.astype(np.float32), y_test.astype(np.int32))
```

```python
#model = L.Classifier(MLP(638,62))
#model = L.Classifier(MLP_bnorm(638, 62))
#model = L.Classifier(MLP_bnorm_dropout(638, 62))
#model = L.Classifier(MLP(638,7))
#model = L.Classifier(MLP_bnorm(638, 7))
model = L.Classifier(MLP_bnorm_dropout(1155, 7))
```

```python
optimizer = chainer.optimizers.MomentumSGD(lr=0.01)
#optimizer = chainer.optimizers.MomentumSGD(lr=0.001)
optimizer.setup(model)
```

```python
train_iter = chainer.iterators.SerialIterator(train, 20)
test_iter = chainer.iterators.SerialIterator(test, 20, repeat=False, shuffle=False)

######### 学習用のパラメータ設定
updater = training.StandardUpdater(train_iter, optimizer, device=-1)
trainer = training.Trainer(updater, (30, 'epoch'), out="logs")
trainer.extend(extensions.Evaluator(test_iter, model, device=-1))
trainer.extend(extensions.LogReport())
trainer.extend(extensions.PrintReport(['epoch', 'main/loss', 'validation/main/loss', 'main/accuracy', 'validation/main/accuracy']))
trainer.extend(extensions.ProgressBar())                                                                                          
```

```python
trainer.run()
```

```python
def eval_score(y_true, y_pred):
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    conf_mat = confusion_matrix(y_true, y_pred)
    print("precision : %04f" % prec)
    print("recall    : %04f" % rec)
    print("accuracy  : %04f" % acc)
    print("f1        : %04f" % f1)
    print("confusion_matrix : ")
    print(confusion_matrix(y_true, y_pred))
    return {"1_prec": prec, "2_rec": rec, "3_acc":acc, "4_f1":f1, "5_conf_mat":conf_mat}
```

```python
with chainer.using_config('train', False):
    # 学習
    y_train_pre = F.softmax(model.predictor(x_train.astype(np.float32))).data.argmax(axis=1)
    eval_score(y_train, y_train_pre)
    # テスト
    y_test_pre = F.softmax(model.predictor(x_test.astype(np.float32))).data.argmax(axis=1)
    eval_score(y_test, y_test_pre)
```
